## Inter Annotator Agreement

```python
IAA_id = '2'
with open(os.path.join(indir, f'IAA{IAA_id}', 'assignments_results.json'), 'r') as f:
    data = json.load(f)
data
```

```python
max_sent_id = max([word['sent_id'] for elem in data.values() for word in elem['summary']])
max_sent_id
```

```python
def get_summary_sent_align_clusters(data, sent_id):
    return list(set([align_cluster for word in data['summary'] for align_cluster in word['alignment_id'] if word['sent_id']==sent_id]))
```

```python
def get_doc_words_ids_in_cluster(data, align_cluster):
    return [word['tkn_id'] for word in data['doc'] if len(list(set.intersection(set(word['alignment_id']), set(align_cluster))))>0 and not word['word'] in string.punctuation]
```

```python
def get_aligned_doc_words(data, align_clusters):
    return {str(sent_id): get_doc_words_ids_in_cluster(data, align_clusters[str(sent_id)]) for sent_id in range(max_sent_id+1)}
```

```python
summary_alignment_clusters = {key:{str(sent_id): get_summary_sent_align_clusters(data[key], sent_id) for sent_id in range(max_sent_id+1)} for key,value in data.items()}
doc_alignment_clusters = {key:get_aligned_doc_words(value, summary_alignment_clusters[key]) for key,value in data.items()}
doc_alignment_clusters
```

```python
comparsion_combinations = list(combinations(summary_alignment_clusters.keys(), 2))
comparsion_combinations
```

## Per Sentence

```python
IAA_scores = {}
for combo in comparsion_combinations:
    IAA_scores[f'{combo[0]}-{combo[1]}'] = {}
    for sent_id in range(max_sent_id+1):
        doc_ids1 = doc_alignment_clusters[combo[0]][str(sent_id)]
        doc_ids2 = doc_alignment_clusters[combo[1]][str(sent_id)]
        IAA_scores[f'{combo[0]}-{combo[1]}'][str(sent_id)] = len(set.intersection(set(doc_ids1), set(doc_ids2))) / len(set.union(set(doc_ids1), set(doc_ids2)))
average = sum([elem for pair in IAA_scores.values() for elem in pair.values()]) / len([elem for pair in IAA_scores.values() for elem in pair.values()])
average
```

## Per Document

```python
IAA_scores_docwise = {}
for combo in comparsion_combinations:
    doc_ids1 = [elem for sent_id in range(max_sent_id+1) for elem in doc_alignment_clusters[combo[0]][str(sent_id)]]
    doc_ids2 = [elem for sent_id in range(max_sent_id+1) for elem in doc_alignment_clusters[combo[1]][str(sent_id)]]
    IAA_scores_docwise[f'{combo[0]}-{combo[1]}'] = len(set.intersection(set(doc_ids1), set(doc_ids2))) / len(set.union(set(doc_ids1), set(doc_ids2)))
average = sum([IAA_score for IAA_score in IAA_scores_docwise.values()]) / len([IAA_score for IAA_score in IAA_scores_docwise.values()])
average
```
